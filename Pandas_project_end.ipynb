Cleaning our dataset

```python
data.dropna(inplace=True)      #droping rows with missing values if any
```

```python
data.fillna(method='ffill', inplace=True)     #filling missing values
```

```python
data.drop_duplicates(inplace=True)           #droping duplicate values if any
```

```python
data['Product ID'] = data['Product ID'].astype('category')       #changing product id to category data type
```

Performing any other necessary modifications in our dataset.

```python
data['space']=data['Region'].map({'North':1,'South':2,'West':3,'East':4})  #creating a new column Space
```

```python
data['prefix']=data['Customer ID'].apply(lambda x:x[0:2])        #creating a new column prefix
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y')        #Changing the data type to date
data['Ship Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y')         #Changing the data type to date
```

```python
data['Shipping Duration'] = (data['Ship Date'] - data['Order Date']).dt.days
np.random.seed(0)  # Setting seed for reproducibility
random_duration = np.random.randint(1, 10, size=len(data))  # Generate random integers between 1 and 10
# Add random duration to 'Order Date' to create 'Ship Date'
data['Ship Date'] = data['Order Date'] + pd.to_timedelta(random_duration, unit='D')
# Calculate shipping duration
data['Shipping Duration'] = random_duration #Creating a random sample duration
```

```python
def seas(x):
    y=x.month
    if y in [12,1]:
        return 'Winter'
    elif y in [2,3]:
        return 'Spring'
    elif y in [4,5,6]:
        return 'Summer'
    elif y in [7,8]:
        return 'Monsoon'
    else:
        return 'Autumn'      
#Creating a function seas to create different seasons
```

```python
data['Seasons']=data['Ship Date'].apply(seas)    #applying seas function to make new column seasons
```

1.Hypothesis: Corporate segment customers (identified by prefixes like "DV") generate higher average profits per order compared to other segments.

```python
segm_grouped=data.groupby('Segment')['Profit'].mean().sort_values()
segm_grouped.plot(kind='bar',color=['blue','green','orange'])
plt.title('Average profit by different segment customers')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()
```

So, it contradicts with our hypothesis as, we can find that home office have more profits.

2.Hypothesis:Orders shipped via "Standard Class" have shorter shipping durations compared to "Second Class," impacting customer satisfaction.

```python
shipmode_grp=data.groupby('Ship Mode')['Shipping Duration'].mean()
shipmode_grp.plot(kind='bar',color=['blue','green','orange','purple'])
plt.title('Average duration by different shipping modes')
plt.xlabel('ship mode')
plt.ylabel('shipping duration')
plt.show()
```

Yes,our analysis also support the hypothesis that, standard class has shorter duration than second class.

3.Hypothesis:Furniture products contribute significantly to total sales revenue but have lower profit margins compared to Office Supplies.


```python
prsales_cate=data.groupby('Category').agg({'Sales':'sum','Profit':'mean'})
prsales_cate.plot(kind='bar', secondary_y= 'Sales',color=['green','blue'])
plt.title('Total sales and profit margin by different categories')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

Hence, with the visualisation we can conclude that furniture products contribute significantly to total sales revenue but have lower profit margins compared to Office Supplies.

4.Hypothesis:Jamestown has the highest amount of sales where as independence is 4th highest in terms of sales as compared to other cities.

```python
citysal_grp=data.groupby('City')['Sales'].mean().nlargest(5)
citysal_grp.plot(kind='bar',color=['red','green','purple','orange','blue'])
plt.title('Top 5 cities with highest average sales revenue ')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

Hence, we can conclude that Jamestown has the highest amount of sales where as independence is 4th highest in terms of sales as compared to other cities.

5.Hypothesis:Moonsoon has the lowest recorded profit margin and the margin always declines in the start of year .

```python
seaspr_grp=data.groupby('Seasons')['Profit'].mean()
seaspr_grp.plot(kind='bar',color=['red','green','purple','orange','blue'])
plt.title('Profit margin recorded in different seasons')
plt.xlabel('Seasons')
plt.ylabel('Profit')
plt.show()
```

```python
data['Month'] = data['Order Date'].dt.to_period('M')
sales_by_month = data.groupby('Month')['Profit'].sum()
plt.figure(figsize=(10, 6))
sales_by_month.plot(marker='o')
plt.title('Total Profit by Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

Hence, we can conclude that Moonsoon has the lowest recorded profit margin and the margin always declines in the start of year
